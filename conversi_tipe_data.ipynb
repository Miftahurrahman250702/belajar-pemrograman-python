```python
# tipe data

# tipe data int
a = 1

# tipe data float
b = 1.2

# tipe data complex
c = 1j


# conversi tipe data float ke int
x = float(a)



# conversi tipe data complex ke float
y = complex(b)


# conversi tipe data complex ke string
z = str(c)

print( x, type(x))

print(y, type(y))

print(z, type(z))

# python adalah bahasa pemrograman berorientasi objek.
# sehingga menggunakan kelas untuk untuk mendefinisikan tipe data
# oleh karena itu melakukan konversi tipe data menggunakan fungsi konstruktor



```
